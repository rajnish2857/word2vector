```python
import pandas as pd
```

```python
data=pd.read_csv('/content/all_kindle_review .csv')
```

```python
df=data[['reviewText','rating']]
df.head()
```

```python
# preprocessing cleaning
df['rating']=data['rating'].apply(lambda x:0 if x<3 else 1 ) # positive revie is 1 and negative review is 1
```

```python
df.head()
```

```python
df['rating'].value_counts()
```

```python
df['reviewText']=df['reviewText'].str.lower()
```

```python
# removing special character
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
!pip install bs4
from bs4 import BeautifulSoup
```

```python
df['reviewText']=df['reviewText'].apply(lambda x:re.sub('[^a-z A-z 0-9]+',' ',x)) #remove special characters
df['reviewText']=df['reviewText'].apply(lambda x:" ".join([y for y in x.split() if y not in stopwords.words('english')])) #removing stopwords
df['reviewText']=df['reviewText'].apply(lambda x:re.sub( r'((http|https):\/\/\S+|www\.\S+|\b\w+\.(com|org|net|io|co|in|ly|me|info|biz|gov|edu)\S*)','',str(x))) # removing url
df['reviewText']=df['reviewText'].apply(lambda x:BeautifulSoup(x,'lxml').get_text()) #remove html tags
df['reviewText']=df['reviewText'].apply(lambda x:" ".join(x.split()))   # remove any additional space
```

```python
df['reviewText']
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer=WordNetLemmatizer()
```

```python
def lemmetize_word(text):
  return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
df['reviewText']=df['reviewText'].apply(lambda x:lemmetize_word(x))
```

```python
df.head()
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df['reviewText'],df['rating'],test_size=0.2,random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
bow=CountVectorizer()
x_train_bow=bow.fit_transform(x_train).toarray()
x_test_bow=bow.transform(x_test).toarray()
```

```python
tfidf=TfidfVectorizer()
x_train_tfidf=tfidf.fit_transform(x_train).toarray()
x_test_tfidf=tfidf.transform(x_test).toarray()
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
nb_model=GaussianNB()
nb_model.fit(x_train_bow,y_train)
y_pred_bow=nb_model.predict(x_test_bow)
accuracy_score(y_test,y_pred_bow)
```

```python
nb_mdl=GaussianNB()
nb_mdl.fit(x_train_tfidf,y_train)
y_pred_tfidf=nb_model.predict(x_test_tfidf)
accuracy_score(y_test,y_pred_tfidf)
```
